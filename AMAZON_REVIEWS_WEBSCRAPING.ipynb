### First we would define all functions required for our web scraping

```python
cookie={} # insert request cookies within{}
def getAmazonSearch(search_products):
    url="https://www.amazon.ca/s?k="+search_products
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
def SearchProductID(asin):
    url="https://www.amazon.ca/dp/"+asin
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
def ScrapeReviews(review_link):
    url="https://www.amazon.ca"+review_link
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
search_products=input("What products would you like to search for?")
product_names=[]
response=getAmazonSearch(search_products)
soup=BeautifulSoup(response.content)
for i in soup.findAll("span",{'class':'a-size-base-plus a-color-base a-text-normal'}): # the tag which is common for all the names of products
    product_names.append(i.text) #adding the product names to the list
```

```python
product_names
```

```python
len(product_names)
```

```python
data_asin=[]
for i in soup.findAll("div",{'class':"sg-col-4-of-12 s-result-item s-asin sg-col-4-of-16 sg-col sg-col-4-of-20"}):
    data_asin.append(i['data-asin'])
```

```python
len(data_asin)
```

```python
link=[]
for i in range(len(data_asin)):
    response=SearchProductID(data_asin[i])
    soup=BeautifulSoup(response.content)
    for i in soup.findAll("a",{'data-hook':"see-all-reviews-link-foot"}):
        link.append(i['href'])
```

```python
len(link)
```

```python
reviews=[]
product_link=[]
for j in range(len(link)):
    for k in range(10):
        response=ScrapeReviews(link[j]+'&pageNumber='+str(k))
        soup=BeautifulSoup(response.content)
        for i in soup.findAll("span",{'data-hook':"review-body"}):
            reviews.append(i.text)
            product_link.append(link[j])
```

```python
review_dict={'reviews':reviews}
product_dict ={'Product Link':product_link}
all_reviews = pd.DataFrame.from_dict(review_dict)
product_list = pd.DataFrame.from_dict(product_dict)
```

```python
columnheaders = ['reviews']
all_reviews.columns = columnheaders
for column in all_reviews.columns:
    all_reviews[column] = all_reviews[column].str.replace(r'\n\n', '')
all_reviews.drop(index=0,inplace=True)

all_reviews['Product Links']=product_list['Product Link']
```

```python
all_reviews.to_csv("amazonreviews.csv")
```
